
## 4. 🔑 Κρίσιμος Κόμβος Απόφασης: Σύμβαση;

### Πίνακας Ταχείας Απόφασης

| Κατάσταση | Βήμα 4 | Βήμα 5 | Βήμα 6 | Σημείωση |
|-----------|:------:|:------:|:------:|----------|
| Πλήρης σύμβαση, μηδέν extra | ❌ | ❌ | ❌ | Εξουσιοδοτημένος ★ |
| Σύμβαση + extra κόστος | ❌ | ✅ | ✅ | Μόνο για το extra |
| Εγγύηση κατασκευαστή | ❌ | ❌ | ❌ | Εξουσιοδοτημένος ★ |
| Χωρίς σύμβαση | ✅ | ✅ | ✅ | Πλήρης ροή |
| Εσωτερική ΤΒΙΤ, χωρίς ανταλλακτικά | ❌ | ❌ | ❌ | Απλούστερο σενάριο |
| Εσωτερική ΤΒΙΤ + ανταλλακτικά | ❌ | ✅ | ✅ | |

> ★ **Προτιμάται πάντα ο Εξουσιοδοτημένος** (official service partner κατασκευαστή)

````python
# 🔑 Interactive Decision Tree
# Απάντησε στις ερωτήσεις για να δεις ποια βήματα χρειάζεσαι

def decide_steps():
    print('=' * 60)
    print('  web-Praxis — INTERACTIVE DECISION TREE')
    print('  Βρες ποια βήματα χρειάζεσαι για τη συγκεκριμένη περίπτωση')
    print('=' * 60)

    # Τύπος συμβάντος
    print('\n📋 ΒΗΜΑ 0: Τι τύπου είναι το συμβάν;')
    print('  1. Βλάβη (απροσδόκητη)')
    print('  2. Περιοδικός Έλεγχος (προγραμματισμένος)')
    event = input('Επίλεξε (1/2): ').strip()

    if event == '1':
        step1 = '✅ Βήμα 1: Αναγγελία Βλάβης'
        step2 = '✅ Βήμα 2: Διαχείριση Βλαβών'
        final_doc = 'Δελτία ΕΠΙΣΚΕΥΗΣ + τιμολόγια'
    else:
        step1 = '✅ Βήμα 1: Δημιουργία Κανόνα ΠΕ (εφάπαξ)'
        step2 = '✅ Βήμα 2: Διαχείριση ΠΕ'
        final_doc = 'Δελτία ΣΥΝΤΗΡΗΣΗΣ + τιμολόγια'

    # Εσωτερική ή εξωτερική;
    print('\n🔧 Εσωτερική ή εξωτερική εκτέλεση;')
    print('  1. Εσωτερικά (τεχνικός ΒΙΤ)')
    print('  2. Εξωτερικά (εταιρεία)')
    who = input('Επίλεξε (1/2): ').strip()

    steps = [step1, step2]

    if who == '1':
        steps.append('✅ Βήμα 3α: Ανάθεση σε τεχνικό ΒΙΤ → Ειδοποίηση αυτόματα ✓')
        print('\n🔩 Χρειάστηκαν ανταλλακτικά;')
        parts = input('(ν/ο): ').strip().lower()
        needs_approval = parts in ['ν', 'ναι', 'y', 'yes']
        needs_market = False
    else:
        steps.append('✅ Βήμα 3β: Κλήση Εταιρείας ⚠️ ΔΕΝ ειδοποιεί αυτόματα!')
        print('\n📄 Υπάρχει σύμβαση συντήρησης;')
        contract = input('(ν/ο): ').strip().lower()

        if contract in ['ν', 'ναι', 'y', 'yes']:
            print('💰 Υπάρχει extra κόστος (εκτός σύμβασης);')
            extra = input('(ν/ο): ').strip().lower()
            needs_approval = extra in ['ν', 'ναι', 'y', 'yes']
            needs_market = False
        else:
            steps.append('✅ Βήμα 4: Έρευνα Αγοράς — Συλλογή & scan προσφορών')
            needs_approval = True
            needs_market = True

    if needs_approval:
        doc_name = 'ΠΕ' if event == '2' else 'Επισκευής'
        steps.append(f'✅ Βήμα 5: Αίτημα Χορήγησης Έγκρισης {doc_name} → Εκτύπωση → ΑΔΑ')
        steps.append('✅ Βήμα 6: Έγκριση — Καταχώριση ΑΔΑ')

    print('\n🚚 Χρειάζεται αποστολή μηχανήματος;')
    ship = input('(ν/ο): ').strip().lower()
    if ship in ['ν', 'ναι', 'y', 'yes']:
        steps.append('✅ Βήμα 7: Αποστολή Μηχανήματος → Εκτύπωση εγγράφου αποστολής')
        steps.append('✅ Βήμα 8: Επιστροφή Μηχανήματος')

    steps.append(f'✅ Βήμα 9: Καταγραφή Αποτελεσμάτων → Επισύναψη: {final_doc}')

    print('\n' + '=' * 60)
    print('  📋 ΑΠΟΤΕΛΕΣΜΑ — Βήματα που χρειάζεσαι:')
    print('=' * 60)
    for i, s in enumerate(steps, 1):
        print(f'  {s}')
    print('\n  Σύνολο βημάτων:', len(steps))
    print('=' * 60)

decide_steps()
````


```
1 → 2 → 3β → 9

```

### 🅔 Σενάριο Β-5 — Χωρίς Σύμβαση (Πλήρης Ροή)
**Προϋπόθεση:** Καμία σύμβαση — πλήρης έρευνα αγοράς

```
1 → 2 → 3β → 4 → 5 → 6 → [7] → [8] → 9
```

**Έγγραφα αυτού του σεναρίου:**
1. Έγγραφο Αναγγελίας/Έρευνας Αγοράς (Βήμα 3β)

3. Αίτηση Έγκρισης Επισκευής (Βήμα 5)

5. Τιμολόγια + Δελτία Επισκευής (Βήμα 9)

### 🅕 Σενάριο Β-6 — Αδυναμία Εσωτερικής, Κλήση Εταιρείας
**Προϋπόθεση:** Ξεκίνησε εσωτερικά, αλλά ο τεχνικός δεν μπόρεσε

```
1 → 2 → 3α → [αδυναμία] → 3β → (4 αν χρειαστεί) → 5 → 6 → 9
```
> ✅ Κλήση και Ανάθεση μπορούν να γίνουν **ΠΟΛΛΕΣ ΦΟΡΕΣ** σε **ΟΠΟΙΟΔΗΠΟΤΕ** στάδιο!

## 7. Αναλυτικός Οδηγός Κάθε Βήματος

### Βήμα 1 — Αναγγελία Βλάβης

**Πρόσβαση:**
```
Τρόπος Α: Πλάγιο Μενού → Επισκευές → Αναφορά Βλάβης
Τρόπος Β: Καρτέλα Μηχανήματος → Κουμπί 'Αναφορά Βλάβης'
          ★ Προτιμητέος: τα στοιχεία μηχανήματος αυτόματα
```

| Πεδίο | Αυτόματο; | Υποχρεωτικό; | Σημείωση |
|-------|:---------:|:------------:|----------|
| Μηχάνημα | Από καρτέλα | ✅ | Αναζήτηση: κωδ./μοντέλο/σειριακό |
| **Περιγραφή** | — | ✅ | Το βλέπει το ΤΒΙΤ — γράψε σαφώς |
| Κατάσταση Λειτουργίας | — | ✅ | Default: Εκτός Χρήσης |
| Ημ/νία Καταχώρισης | ✅ | — | ΔΕΝ αλλάζει ποτέ |
| Όνομα Αναγγέλλοντος | ✅ Από login | — | |

**Μετά Αποθήκευση:**
```diff
+ Αυτόματος Κωδικός Βλάβης
+ Αυτόματη ειδοποίηση στο ΤΒΙΤ
+ Κατάσταση: 'Νέα Αναγγελία' 🟢
```

---

### Βήμα 3α — Ανάθεση σε ΤΒΙΤ

## 11. 🚀 Interactive Decision Tree

> Τρέξε το κελί κώδικα παρακάτω για να χρησιμοποιήσεις τον interactive οδηγό.
> Απάντησε στις ερωτήσεις και θα σου πει ακριβώς ποια βήματα χρειάζεσαι.

````python
# 🚀 INTERACTIVE DECISION TREE — web-Praxis
# Τρέξε και απάντησε στις ερωτήσεις!

def web_praxis_wizard():
    print('\n' + '🏥 ' * 20)
    print('  web-Praxis WIZARD — Βρες τα σωστά βήματα')
    print('🏥 ' * 20)

    steps = []
    notes = []

    # Step 0: Τύπος
    print('\n❓ Τι τύπου είναι το συμβάν;')
    print('   [1] Βλάβη (απροσδόκητη)')
    print('   [2] Περιοδικός Έλεγχος (προγραμματισμένος)')
    t = input('→ ').strip()
    is_vlavi = (t != '2')

    if is_vlavi:
        steps.append(('1', 'Αναγγελία Βλάβης', 'Φόρμα + αυτόματος κωδικός + ειδοποίηση ΤΒΙΤ'))
        steps.append(('2', 'Διαχείριση Βλαβών', 'Ορισμός Υπεύθυνου Βλάβης'))
        final_attach = 'τιμολόγια + δελτία ΕΠΙΣΚΕΥΗΣ'
    else:
        steps.append(('1', 'Δημιουργία Κανόνα ΠΕ', 'Εφάπαξ ρύθμιση · Έγγραφο ΟΔΙΠΥ (προαιρετικό)'))
        steps.append(('2', 'Διαχείριση ΠΕ', 'Ορισμός υπεύθυνου'))
        final_attach = 'τιμολόγια + δελτία ΣΥΝΤΗΡΗΣΗΣ'

    # Step 3: Εσωτερικά ή εξωτερικά
    print('\n❓ Εσωτερική ή εξωτερική εκτέλεση;')
    print('   [1] Εσωτερικά — τεχνικός ΒΙΤ')
    print('   [2] Εξωτερικά — εταιρεία')
    who = input('→ ').strip()

    needs_approval = False

    if who == '1':
        steps.append(('3α', 'Ανάθεση σε τεχνικό ΒΙΤ',
                      'Ειδοποίηση τεχνικού ΑΥΤΟΜΑΤΑ ✓ · Κατάσταση → Εκκρεμότητα 🟠'))
        print('\n❓ Χρειάστηκαν ανταλλακτικά;')
        print('   [ν] Ναι   [ο] Όχι')
        parts = input('→ ').strip().lower()
        if parts in ['ν', 'ναι', 'y']:
            needs_approval = True
    else:
        steps.append(('3β', 'Κλήση Εταιρείας',
                      '⚠️ ΔΕΝ ειδοποιεί αυτόματα! Τηλεφώνησε εσωτερικά · Κατάσταση → Εκκρεμότητα 🟠'))
        notes.append('📄 Εκτύπωσε έγγραφο Αναγγελίας/Έρευνας Αγοράς')

        print('\n❓ Υπάρχει σύμβαση συντήρησης;')
        print('   [ν] Ναι   [ο] Όχι')
        contract = input('→ ').strip().lower()

        if contract in ['ν', 'ναι', 'y']:
            print('\n❓ Υπάρχει extra κόστος (εκτός σύμβασης);')
            print('   [ν] Ναι   [ο] Όχι')
            extra = input('→ ').strip().lower()
            needs_approval = extra in ['ν', 'ναι', 'y']
            if not needs_approval:
                notes.append('✅ Πλήρης σύμβαση → Παράλειψε Βήματα 4, 5, 6')
        else:
            steps.append(('4', 'Έρευνα Αγοράς',
                          'Συλλογή προσφορών · Scan/φωτογραφία · Επισύναψη ΟΛΩΝ'))
            needs_approval = True

    if needs_approval:
        doc = 'ΠΕ' if not is_vlavi else 'Επισκευής'
        steps.append(('5', f'Αίτημα Χορήγησης Έγκρισης {doc}',
                      f'Εκτύπωση εγγράφου · Υπογραφές → ΑΔΑ · Κατάσταση → Αν.Έγκρισης 🔵'))
        steps.append(('6', 'Έγκριση',
                      'Καταχώριση ΑΔΑ · Κατάσταση → Αναμονή Επισκευής 🩷'))

    print('\n❓ Χρειάζεται αποστολή μηχανήματος εκτός νοσοκομείου;')
    print('   [ν] Ναι   [ο] Όχι')
    ship = input('→ ').strip().lower()
    if ship in ['ν', 'ναι', 'y']:
        steps.append(('7', 'Αποστολή Μηχανήματος', 'Εκτύπωση εγγράφου αποστολής'))
        steps.append(('8', 'Επιστροφή Μηχανήματος', 'Καταγραφή ημερομηνίας επιστροφής'))

    steps.append(('9', 'Καταγραφή Αποτελεσμάτων',
                  f'ΠΑΝΤΑ ΥΠΟΧΡΕΩΤΙΚΟ · Επισύναψη: {final_attach} · Κατάσταση → Ολοκληρωμένη ⬜'))

    # Output
    print('\n' + '=' * 65)
    print(f'  ✅ ΠΛΑΝΟ ΕΝΕΡΓΕΙΩΝ — {len(steps)} βήματα')
    print('=' * 65)
    for step, title, detail in steps:
        print(f'\n  📌 Βήμα {step}: {title}')
        print(f'     {detail}')

    if notes:
        print('\n' + '-' * 65)
        print('  📝 Σημειώσεις:')
        for n in notes:
            print(f'  {n}')

    print('\n' + '=' * 65)
    print('  ⭐ Προτιμάται πάντα ο ΕΞΟΥΣΙΟΔΟΤΗΜΕΝΟΣ service partner!')
    print('=' * 65)

web_praxis_wizard()
````
